```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def cem(func, x_min=0, x_max=2, dim=1, iterMax=10, sampleNum=700, eliteNum=200):

    mean = None
    std = None
    sampleSet = None
    valueSet = None

    for i in range(iterMax):
        # 正規分布に従うサンプルを抽出
        if i ==0:
            sampleSet_i = np.random.uniform(x_min, x_max, (sampleNum, dim))
        else:
            sampleSet_i = np.random.normal(loc=mean, scale=std, size=sampleNum)
        
        # 正規分布から得られたサンプル
        sampleSet = np.reshape(sampleSet_i, (sampleNum, -1))
        # 正規分布から得られたサンプルを真の分布に代入
        valueSet = func(sampleSet)
        # 関数ｆuncの出力結果のなかで、エリートを選出
        value_sort = np.sort(valueSet, axis=None)
        data_num = sampleSet.shape[0]
        # エリートのビリ
        v_cut = value_sort[data_num - eliteNum -1]
        #print('v_cut = ', v_cut)
        idx = (valueSet.flatten() > v_cut)
        
        # elite value
        valueSet_elite = valueSet[idx]
        # エリートになるときのサンプル入力
        sampleSet_elite = sampleSet[idx]

        # エリートになるときのサンプル入力から作られる正規分布を次の世代に引き継がせる
        # 交差エントロピー法では、f最大になるときのｘが平均で標準偏差が０の正規分布（真の分布）に、サンプル元の仮の正規分布qを近づけるような最適化が行われる。
        
        mean = np.mean(sampleSet_elite)
        std = np.std(sampleSet_elite)

    return mean

def func(x):
    return x*np.sin(x)
```

```python
x_min = 0
x_max = 10

max_x = cem(func, x_min, x_max, iterMax=10, sampleNum=100, eliteNum=40) # sampleSet_i
print('max_x = ', max_x)

x = np.linspace(x_min, x_max, 50)
y = func(x)
plt.plot(x, y)
plt.vlines(max_x, -6, 9, color='r', linestyles='dotted')
```

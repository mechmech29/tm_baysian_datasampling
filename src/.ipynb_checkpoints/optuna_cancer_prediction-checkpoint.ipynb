```python
import optuna
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.svm
from sklearn import linear_model # linear_model.LogisticRegression ロジスティック回帰
from sklearn import ensemble # アンサンブル学習
from sklearn import datasets # 乳がんデータ
from sklearn import model_selection

#from .autonotebook import tqdm as notebook_tqdm

# disable optuna logging
def optuna_log(log=False):
    if log:
        optuna.logging.enable_default_handler()
    else:
        optuna.logging.disable_default_handler()

optuna_log(False)
```

```python
# objective function
def f(x):
    return x*np.sin(x)

# objective function for optuna optimizer
def objective(trial): # optuna.trial, ...
    name = 'x'
    low, high = -7, 13
    # set parameter for optimization
    x = trial.suggest_float('x', low, high)
    return f(x)

# optimize the objective function
study = optuna.create_study(direction='maximize') #(direction='minimize') # instance optimizer
study.optimize(objective, n_trials=100, show_progress_bar=False) # optimize
print('optimization is done !')
```

```python
print('best objective value in study = %.3f'%study.best_value)

print('best objective param x in study = ', study.best_params)

print(np.max(f(np.linspace(-7, 13,100))))
print(f(study.best_params['x']))

x_plt = np.linspace(-7, 13, 100)
plt.plot(x_plt, f(x_plt))
```

定番？の乳がんデータの分類問題をoptunaによるハイパーパラメータ最適化で解いてみる  
ロジスティック回帰，サポートベクターマシン（SVM），ランダムフォレストで分類  
それぞれのモデルにハイパーパラメータを設定して最適化

```python
# load sample data
X_data, y_data = datasets.load_breast_cancer(return_X_y=True, as_frame=True)
pd.DataFrame(X_data).head()
```

```python
# define objective fuction
def objective(trial):
    # 分類器の名前
    #classifier_name = "LogReg" #trial.suggest_categorical("classifier", ["LogReg", "SVC", "RandomForest"])
    classifier_name = "SVC"
    range = (1e-10, 1e10)

    # なぜかロジスティック回帰だけがうまくいかない．．．
    if classifier_name == "LogReg": # ロジスティック回帰
        param = trial.suggest_float("logreg_c", range[0], range[1], log=True)
        classifier_obj = linear_model.LogisticRegression(C=param) # C is a parameter which controls strictness

    elif classifier_name == "SVC": # サポートベクターマシン
        param = trial.suggest_float("svc_param", range[0], range[1], log=True)
        classifier_obj = sklearn.svm.SVC(C=param, gamma="auto")

    #else classifier_name == "RandomForest": # ランダムフォレスト
    else:
        #param = trial.suggest_trial("rndfost_pram", range[0], range[1], log=True)
        #classifier_obj = 
        rf_n_estimators = trial.suggest_int("rf_n_estimators", 10, 1000)
        rf_max_depth = trial.suggest_int("rf_max_depth", 2, 50, log=True)
        classifier_obj = ensemble.RandomForestClassifier(
            max_depth=rf_max_depth, n_estimators=rf_n_estimators
        )

    # 正答率を出す
    # cross validation score 交差検証
    score = model_selection.cross_val_score(classifier_obj, X_data, y_data, n_jobs=-1, cv=10)
    accuracy = score.mean()
    return accuracy

def objective_simple(trial):
    #param = trial.suggest_float("logreg_c", 1e-5, 1e5, log=True)
    #classifier_obj = linear_model.LogisticRegression(C=param)
    rf_n_estimators = trial.suggest_int("rf_n_estimators", 10, 1000)
    rf_max_depth = trial.suggest_int("rf_max_depth", 2, 50, log=True)
    classifier_obj = ensemble.RandomForestClassifier(
        max_depth=rf_max_depth, n_estimators=rf_n_estimators
    )   

    score = model_selection.cross_val_score(classifier_obj, X_data, y_data, n_jobs=-1, cv=10)
    accuracy = score.mean()
    return accuracy

optuna_log(False)
study = optuna.create_study(direction='maximize')
study.optimize(objective_simple, n_trials=10)
#study.optimize(objective, n_trials=10)
print('optimization is finished !')
```

```python
print('optimized value = ', study.best_value)
print('optimized param = ', study.best_params)
```

ロジスティック回帰，SVC，ランダムフォレストのうち，最も高い正答率を出せたのはランダムフォレスト  
経緯を`optuna.visualization`で描画することができる．


# ガウス過程回帰の練習

ガウス過程回帰を使うプログラムです。
以下のページのプログラムを用いていますが，本レクチャーに合わせた解説をこのプログラムでは載せています。
https://scikit-learn.org/stable/auto_examples/gaussian_process/plot_gpr_noisy_targets.html

このプログラムでは，出力が1次元で，以下の２つの例を示しています。
1. 観測にノイズがない場合
2. 観測にノイズ(ノイズの程度は既知)がある場合

実験系におけるベイズ最適化では，観測にノイズがある場合がほとんどだと思いますので，二つ目の例が主になると思います。
どちらの場合でも，尤度の最大化によってカーネル関数のパラメータを推定します。

図では，95%信頼区間を表示することで，推定の信頼度を可視化できるガウス過程回帰の特徴を見てもらえると思います。

 `alpha` はTikhonov正規化とよばれるテクニックのパラメータになっています。

## 人工的なデータの生成

ここでは，推定する未知の関数として
 $f(x) = x \sin(x)$を使います。



```python
import numpy as np

X = np.linspace(start=0, stop=10, num=1_000).reshape(-1, 1)
y = np.squeeze(X * np.sin(X))
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y, label=r"$f(x) = x \sin(x)$", linestyle="dotted")
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
_ = plt.title("True generative process")
```

この点線が，未知の関数の真の値を示しています。
以下では，この関数をガウス過程回帰で推定します。

## 観測にノイズがない場合

以下では，観測にノイズがない場合を示します。
例では，６個の観測データを与えた場合を考えます。



```python
rng = np.random.RandomState(1)
training_indices = rng.choice(np.arange(y.size), size=6, replace=False)
X_train, y_train = X[training_indices], y[training_indices]
```

次に，ガウス過程回帰を行います。
カーネル関数としては，放射基底関数(radial basis function, RBF)を用います。



```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gaussian_process = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
gaussian_process.fit(X_train, y_train)
gaussian_process.kernel_
```

gaussian_process.fit()を実行すると，カーネル関数のパラメータが最適化されます。
この最適化されたカーネル関数を用いて推定を行い，95%信頼区間を表示します。



```python
mean_prediction, std_prediction = gaussian_process.predict(X, return_std=True)

plt.plot(X, y, label=r"$f(x) = x \sin(x)$", linestyle="dotted")
plt.scatter(X_train, y_train, label="Observations")
plt.plot(X, mean_prediction, label="Mean prediction")
plt.fill_between(
    X.ravel(),
    mean_prediction - 1.96 * std_prediction,
    mean_prediction + 1.96 * std_prediction,
    alpha=0.5,
    label=r"95% confidence interval",
)
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
_ = plt.title("Gaussian process regression on noise-free dataset")
```

信頼区間を可視化してみると，観測データから離れた点では，推定値の分散が大きくなり，観測データから近いところでは推定値の分散が小さくなることが見て取れます。
また，観測データにはノイズがのっていない，という仮定を用いたモデルのため，ガウス過程回帰で推定した平均の線は観測データの点を通り，観測データの点上では推定の分散がゼロになります。

## 観測データにノイズがのっている場合

次に，観測データにノイズがのっている場合を扱います。
以下のようにして，観測データにノイズを加えてみます。



```python
noise_std = 0.75
y_train_noisy = y_train + rng.normal(loc=0.0, scale=noise_std, size=y_train.shape)
```

次に，ガウス過程回帰モデルを用意します。
カーネル関数に加えて，パラメータ`alpha` を設定します。
これは，観測に乗るノイズの分散と解釈することができます。




```python
gaussian_process = GaussianProcessRegressor(
    kernel=kernel, alpha=noise_std**2, n_restarts_optimizer=9
)
gaussian_process.fit(X_train, y_train_noisy)
mean_prediction, std_prediction = gaussian_process.predict(X, return_std=True)
```

カーネル関数をgaussian_proceess.fit()で最適化した後，さきほどと同じようにグラフを表示してみます。



```python
plt.plot(X, y, label=r"$f(x) = x \sin(x)$", linestyle="dotted")
plt.errorbar(
    X_train,
    y_train_noisy,
    noise_std,
    linestyle="None",
    color="tab:blue",
    marker=".",
    markersize=10,
    label="Observations",
)
plt.plot(X, mean_prediction, label="Mean prediction")
plt.fill_between(
    X.ravel(),
    mean_prediction - 1.96 * std_prediction,
    mean_prediction + 1.96 * std_prediction,
    color="tab:orange",
    alpha=0.5,
    label=r"95% confidence interval",
)
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
_ = plt.title("Gaussian process regression on a noisy dataset")
```

観測データにノイズがのっているので，ガウス過程回帰で推定される平均の線は，観測データを通らず，観測データがある点においても，推定の分散はゼロになりません。
観測データから遠くなると推定の分散が大きくなり，観測データに近くなると推定の分散が小さくなるという傾向は，前の例と同じになります。
